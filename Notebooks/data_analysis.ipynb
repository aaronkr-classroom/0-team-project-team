# EDA 분석

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='AppleGothic') 			
plt.rcParams['axes.unicode_minus'] = False 

# Load data
data = pd.read_csv("../Data/train.csv")

plt.figure(figsize=(15, 20))

# person_age
plt.subplot(3, 2, 1)
plt.hist(data['person_age'], bins=20, edgecolor='k', alpha=0.7)
plt.title('Distribution of Person Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

# person_income
plt.subplot(3, 2, 2)
plt.hist(data['person_income'], bins=20, edgecolor='k', alpha=0.7)
plt.title('Distribution of Person Income')
plt.xlabel('Income')
plt.ylabel('Frequency')

# Plot 3: Loan amount by loan grade
plt.subplot(3, 2, 3)
data.groupby('loan_grade')['loan_amnt'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Loan Amount by Loan Grade')
plt.xlabel('Loan Grade')
plt.ylabel('Average Loan Amount')

# Plot 4: Average interest rate by loan intent
plt.subplot(3, 2, 4)
data.groupby('loan_intent')['loan_int_rate'].mean().plot(kind='bar', color='coral')
plt.title('Average Interest Rate by Loan Intent')
plt.xlabel('Loan Intent')
plt.ylabel('Average Interest Rate (%)')

# Plot 5: Loan Percent Income by Credit History Length
plt.subplot(3, 2, 5)
plt.scatter(data['cb_person_cred_hist_length'], data['loan_percent_income'], alpha=0.5)
plt.title('Loan Percent Income by Credit History Length')
plt.xlabel('Credit History Length (years)')
plt.ylabel('Loan Percent of Income')

# Plot 6: Count of Home Ownership Types
plt.subplot(3, 2, 6)
data['person_home_ownership'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Count of Home Ownership Types')
plt.xlabel('Home Ownership')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
